```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
df = pd.read_csv("blogData_train.csv")
df.head()
```

```python
X = df.iloc[:,:181]
y = df["1.0.2"]
X.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# #############################################################################
# Generate some sparse data to play with
np.random.seed(42)

#n_samples, n_features = 14723276, 181
#coef = 3 * np.random.randn(n_features)
#inds = np.arange(n_features)
#coef[inds[10:]] = 0  # sparsify coef

# add noise
#y += 0.01 * np.random.normal(size=n_samples)

# Split data in train set and test set
n_samples = X.shape[0]
X_train, y_train = X[:n_samples // 2], y[:n_samples // 2]
X_test, y_test = X[n_samples // 2:], y[n_samples // 2:]
```

```python
# ElasticNet
from sklearn.linear_model import ElasticNetCV

enet = ElasticNetCV(alphas=np.linspace(1,10,5), l1_ratio=np.linspace(0.1,1,5))
```

```python
print(X_train.shape)
print(y_train.shape)
y_pred_enet = enet.fit(X_train, y_train)
print(y_pred_enet)
y_pred_enet = y_pred_enet.predict(X_test)
print(y_pred_enet)
r2_score_enet = r2_score(y_test, y_pred_enet)
```

```python

print(enet)

print("r^2 on test data : %f" % r2_score_enet)

plt.plot(enet.coef_, color='lightgreen', linewidth=2,
         label='Elastic net coefficients')
plt.legend(loc='best')
plt.show()
```
